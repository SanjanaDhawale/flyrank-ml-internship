## Target or proxy

The target for this project is whether a content page is likely to experience a decline in performance. In the starter dataset, the proxy target is the `trend_direction` column. Pages with a value of **"down"** can be treated as declining, while the remaining pages represent stable or improving performance.

This target is derived from observed historical performance trends in the dataset rather than being manually assigned. It provides a practical label for training and evaluating a classification model.

## Success metric

The primary success metric for this classification task is **F1-score** because it balances both precision and recall. This is important since we want to correctly identify declining content pages while minimizing false positives and false negatives.

A higher F1-score indicates that the model is effectively identifying pages that require content refresh without recommending unnecessary updates.

## Why ML beats a fixed rule here

A fixed rule cannot effectively capture the complex relationships between search volume, user engagement, content age, traffic trends, and other factors that influence content performance.

Machine learning can learn patterns from multiple features simultaneously and identify declining content more accurately than a simple rule such as "refresh every page older than six months." This enables content teams to prioritize pages based on data-driven predictions instead of relying on manual or fixed thresholds.
