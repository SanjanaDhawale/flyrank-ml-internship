## My Rule

A content page should be reviewed for refresh if it has high search visibility, has not been updated recently, and has a low click-through rate (CTR). These signals suggest that the page is receiving impressions but is not attracting enough clicks and may benefit from refreshed content.

### Reason Codes

- STALE_CONTENT
- LOW_CTR
- HIGH_IMPRESSIONS

### Action Label

REFRESH_CONTENT

## Build the Ranked Queue

The baseline score is built using simple, transparent rules instead of machine learning.

Scoring Logic:
- High impressions indicate that the page has good visibility.
- Older content is more likely to need refreshing.
- Low CTR suggests users see the page but are not clicking it.

Each page receives:
- A baseline score
- A reason code
- An action label

The ranked queue is sorted by the baseline score in descending order and saved as `work/outputs/baseline_action_score.csv`.

```python
import os

# Create output folder if it doesn't exist
os.makedirs("work/outputs", exist_ok=True)

# Baseline score
df["baseline_score"] = (
    (df["impressions_90d"] / 100) +
    (df["days_since_last_update"] / 30) +
    ((1 - df["ctr"]) * 20)
)

# Reason code
def reason(row):
    reasons = []

    if row["days_since_last_update"] >= 180:
        reasons.append("STALE_CONTENT")

    if row["impressions_90d"] >= 500:
        reasons.append("HIGH_IMPRESSIONS")

    if row["ctr"] <= 1:
        reasons.append("LOW_CTR")

    if len(reasons) == 0:
        return "NONE"

    return ",".join(reasons)

df["reason_code"] = df.apply(reason, axis=1)

# Action label
df["action"] = df["baseline_score"].apply(
    lambda x: "REFRESH_CONTENT" if x > 20 else "NO_ACTION"
)

# Rank pages
ranked = df.sort_values(
    by="baseline_score",
    ascending=False
)

# Save CSV
ranked.to_csv(
    "work/outputs/baseline_action_score.csv",
    index=False
)

print("CSV created successfully!")
print(ranked[[
    "baseline_score",
    "reason_code",
    "action"
]].head(10))
```

## Weak Picks

Some high-ranked pages may not actually require a content refresh. For example, pages with high impressions but naturally low CTR (such as informational queries or SERP-feature-heavy searches) may appear in the ranked queue even though the content itself is performing as expected. Likewise, seasonal trends or temporary ranking fluctuations may reduce CTR without indicating poor content quality.

These cases should be manually reviewed before taking action. The baseline is intended as a decision-support tool rather than an automated decision maker.

## Leakage Check

The baseline score was built using only input features available before making the recommendation.

The following columns were **not** used because they represent future outcomes or label-derived information:

- `trend_direction`
- `trend_pct`

No product-generated flags, future-window information, or label-derived features were used while calculating the baseline score. Therefore, the baseline does not introduce data leakage.

```python
print("Weak Picks Review")
print("-" * 50)

weak_picks = ranked[
    (ranked["impressions_90d"] > 1000) &
    (ranked["ctr"] > 2)
].head(10)

print(weak_picks[[
    "content_id",
    "impressions_90d",
    "ctr",
    "baseline_score",
    "reason_code"
]])

print("\nLeakage Check")
print("-" * 50)
print("✓ trend_direction not used")
print("✓ trend_pct not used")
print("✓ No future-window features used")
print("✓ No label-derived features used")
```
